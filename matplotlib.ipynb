```python
import matplotlib as mpl
import matplotlib.pyplot as plt
plt.style.use('ggplot')

```

```python
#basic
#For all matplotlib plots, first create a figure and an axes object; to show the plot, call “plt.show()”. The figure contains all the objects, including axes, graphics, texts, and labels. The axes is a bounding box with ticks and labels. Think of axes as an individual plot.

#matplotlib.pyplot provides many customization features as we will see in the following lessons.

```

```python
#example
import numpy as np
x = np.linspace(0, 10, 1000)
#print(x)
y = np.sin(x)
fig = plt.figure()
ax = plt.axes()

ax.plot(x, y)
plt.show()
```

```python
#Labels and Titles
x = np.linspace(0, 10, 1000) #this number between 0 and 10
y = np.cos(x) 

plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("function cos(x)")
plt.show()

#One can also set the limits of x- and y-axis using plt.xlim() and plt.ylim(), respectively.
print("-------------------------------------------------------")
plt.plot(x, y)
plt.xlabel("this is x")
plt.ylabel("this is ay")
plt.title("function cos(x)")
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.show()
```

```python
#Multiple Lines part1
x = np.linspace(0, 10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1)
plt.plot(x, y2)
plt.show()
```

```python
#Multiple Lines part2
x = np.linspace(0, 10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1, color = "r", linestyle = "-")
plt.plot(x, y2, color = "g", linestyle = "--")
plt.show()
```

```python
x = np.linspace(0,10,1000) # 1darray of length 1000
plt.plot(x, np.sin(x), 'k:', label='sin(x)')
plt.plot(x, np.cos(x), 'r--', label='cos(x)')
plt.legend()
plt.savefig("plot.png")
plt.show()
```

```python
#Scatter Plot
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

presidents_df = pd.read_csv('https://sololearn.com/uploads/files/president_heights_party.csv', index_col='name')

plt.scatter(presidents_df['height'], presidents_df['age'])
plt.show()
```

```python
#label in scatter
plt.scatter(presidents_df['height'], presidents_df['age'], marker="<", color = "b")
plt.xlabel('height')
plt.ylabel('age')
plt.title("U.S. president")
plt.show()
```

```python
#Plotting with Pandas
presidents_df.plot(kind='scatter',
                   x = 'height',
                   y = 'age',
                   title = 'U.S. President')
plt.show()
```

```python
#histogram part1
presidents_df['height'].plot(kind='hist',
                             title="height",
                             bins=5)
plt.show()
```

```python
#histogram part2
plt.hist(presidents_df['height'], bins=5)
plt.show()
```

```python
#boxplot part1

plt.style.use("classic")
presidents_df.boxplot(column='height')
plt.show()
```

```python
#box plot part2
plt.boxplot(presidents_df['height'])
plt.show()
```

```python
#barplot part1
party_cnt = presidents_df['party'].value_counts()
party_cnt.plot(kind="bar")
plt.show()
```

```python
#bar plot part2
part_cnt = presidents_df['party'].value_counts()
plt.bar(part_cnt)
plt.show()
```
